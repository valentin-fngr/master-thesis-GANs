```python
import numpy as np 
import matplotlib.pyplot as plt 
import torch.nn as nn 
import torch.optim as optim 
import torchvision.transforms as transforms
import torchvision.datasets as datasets
import torch

import torchvision
```

```python
IMG_FOLDER = "data"
```

### hyperparameters

```python
batch_size = 128 
learning_rate = 0.0002
beta_momentum = 0.5
gen_out_shape = 64
```

```python
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
print("device : ", device)
```

```python
transform = transforms.Compose([
    transforms.Resize(gen_out_shape),
    transforms.ToTensor(), 
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

data_loader = datasets.ImageFolder(IMG_FOLDER,transform=transform)
print("dataset size : ", len(data_loader))
```

```python
data_loader = torch.utils.data.DataLoader(data_loader,
                                          batch_size=batch_size,
                                          shuffle=True,
                                          num_workers=2,
                                          drop_last=True
                                         )
```

```python
data = iter(data_loader)
images, label = data.next()
```

```python
print(label)
grid = torchvision.utils.make_grid(images[:10])
```

```python
def plot_image(inp, title=None): 
    inp = np.transpose(inp, axes=[1, 2, 0])
    plt.title(title)
    plt.imshow(inp)
    return
```

```python
plot_image(grid)
```

```python
def custom_init(m): 
    if isinstance(m, nn.ConvTranspose2d) or isinstance(m, nn.Con2d) or isinstance(m, nn.BatchNorm2d): 
        nn.init.normal_(m.weight, mean=0.0, std=0.02)
```

```python
import torch.nn.functional as F_

class Generator(nn.Module): 
    
    def __init__(self, ns, output_size): 
        super(Generator, self).__init__()
        self.ns = ns 
        self.output_size = output_size
        self.upconv1 = nn.ConvTranspose2d(ns, 1024, 4)
        self.bn1 = nn.BatchNorm2d(1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 4, 2, 1)
        self.bn2 = nn.BatchNorm2d(512)
        self.upconv3 = nn.ConvTranspose2d(512, 256, 4, 2, 1)
        self.bn3 = nn.BatchNorm2d(256)
        self.upconv4 = nn.ConvTranspose2d(256, 128, 4, 2, 1)
        self.bn4 = nn.BatchNorm2d(128)
        self.upconv5 = nn.ConvTranspose2d(128, 3, 4, 2, 1)
        self.bn5 = nn.BatchNorm2d(3)
    
    def forward(self, x): 
        
        x = F_.relu(self.bn1(self.upconv1(x)))
        print(x.shape)
        x = F_.relu(self.bn2(self.upconv2(x)))
        print(x.shape)
        x = F_.relu(self.bn3(self.upconv3(x)))
        print(x.shape)
        x = F_.relu(self.bn4(self.upconv4(x)))
        print(x.shape)
        x = F_.relu(self.bn5(self.upconv5(x)))
        print(x.shape)
        
        return x
```

```python
# summary 
from torchsummary import summary

generator = Generator(100, gen_out_shape)
generator.to(device)
summary(generator, input_size=(100,1, 1), batch_size=1, device="cpu")
```
